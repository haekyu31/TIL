```python
# This time we are going to pull data directly from the internet.
# Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

### Step 1. Import the necessary libraries
import numpy as np
import pandas as pd
```

```python
### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
```

```python
### Step 3. Assign it to a variable called chipo.
chipo = pd.DataFrame(chipo)
chipo
```

```python
### Step 4. $10.00 이상인 제품이 몇 개 있습니까?
def price(x):
    x_list = x.split("$")
    return float(x_list[1])

chipo.item_price = chipo.item_price.apply(price)
```

```python
# a = chipo["item_price"]>10  #10달러 이상인 주문
# chipo[a].count
chipo[["item_name", "item_price"]].groupby(by = "item_name").agg(["min","max"])
```

```python
# chipo[["item_name", "item_price"]].groupby(by = "item_name").agg(["min","max"]).columns
boolprice = chipo[["item_name", "item_price"]].groupby(by = "item_name").agg(["min","max"])[("item_price","max"])>10
```

```python
### Step 4. $10.00 이상인 제품이 몇 개 있습니까?
def price(x):
    x_list = x.split("$")
    return x_list[1]
```

```python
### Step 5. 각 품목의 가격은 얼마입니까?
###### item_name 과 item_price 만 출력하시오
chipo[["item_name","item_price"]].value_counts()
```

```python
chipo[["item_name","item_price"]].groupby(by = "item_name").agg(["min","max"])
```

```python
### Step 6. item_name으로 정렬하세요
chipo.sort_values(by = "item_name")
chipo.item_name.sort_values()
```

```python
### Step 7. 가장 비싼 품목의 주문 수량은 몇개입니까?
chipo.sort_values(by = "item_price", ascending = False).head(1)
```

```python
### Step 8. Veggie Salad Bowl은 몇번 주문 되었습니까?
# chipo.groupby(by = "item_name").count()
# a1 = chipo.groupby(by = "item_name")

# chipo.item_name.value_counts()["Veggie Salad Bowl"]
chipo["item_name"].value_counts()["Veggie Salad Bowl"]
```

```python
### Step 9. Canned Soda를 두 개 이상 주문한 횟수는 몇번입니까?
bool_soda = (chipo.item_name =='Canned Soda') & (chipo.quantity >1)
```
